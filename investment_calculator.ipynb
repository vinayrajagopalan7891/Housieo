Mortgage Calculator

```python
import sys
import argparse
import decimal

MONTHS_IN_YEAR = 12
DOLLAR_QUANTIZE = decimal.Decimal('.01')

def dollar(f, round=decimal.ROUND_CEILING):
    """
    This function rounds the passed float to 2 decimal places.
    """
    if not isinstance(f, decimal.Decimal):
        f = decimal.Decimal(str(f))
    return f.quantize(DOLLAR_QUANTIZE, rounding=round)

class Mortgage:
    def __init__(self, interest, months, amount):
        self._interest = float(interest)
        self._months = int(months)
        self._amount = dollar(amount)

    def rate(self):
        return self._interest

    def month_growth(self):
        return 1. + self._interest / MONTHS_IN_YEAR

    def apy(self):
        return self.month_growth() ** MONTHS_IN_YEAR - 1

    def loan_years(self):
        return float(self._months) / MONTHS_IN_YEAR

    def loan_months(self):
        return self._months

    def amount(self):
        return self._amount

    def monthly_payment(self):
        pre_amt = float(self.amount()) * self.rate() / (float(MONTHS_IN_YEAR) * (1.-(1./self.month_growth()) ** self.loan_months()))
        return dollar(pre_amt, round=decimal.ROUND_CEILING)

    def total_value(self, m_payment):
        return m_payment / self.rate() * (float(MONTHS_IN_YEAR) * (1.-(1./self.month_growth()) ** self.loan_months()))

    def annual_payment(self):
        return self.monthly_payment() * MONTHS_IN_YEAR

    def total_payout(self):
        return self.monthly_payment() * self.loan_months()

    def monthly_payment_schedule(self):
        monthly = self.monthly_payment()
        balance = dollar(self.amount())
        rate = decimal.Decimal(str(self.rate())).quantize(decimal.Decimal('.000001'))
        while True:
            interest_unrounded = balance * rate * decimal.Decimal(1)/MONTHS_IN_YEAR
            interest = dollar(interest_unrounded, round=decimal.ROUND_HALF_UP)
            if monthly >= balance + interest:
                yield balance, interest
                break
            principle = monthly - interest
            yield principle, interest
            balance -= principle

# Source https://github.com/jbmohler/mortgage/blob/master/mortgage.py
```

Total Payout

```python
listings_final['mortgage_payout'] = listings_final.apply(lambda x: Mortgage(interest=x.mortgage_rate, amount=x.mortgage_amount, months=x.mortgage_duration).total_payout(), axis=1)
listings_final['mortgage_payout'] = listings_final['mortgage_payout'].astype(int)
listings_final.head()
```

Total Interest Paid

```python
listings_final['interest_paid'] = listings_final['mortgage_payout'] - listings_final['mortgage_amount']
listings_final.head()
```

Mortgage recalculated after subracting the listing price

```python
listings_final['mortgage_amount_after_listing'] = listings_final['mortgage_amount'] - listings_final['listing_amount']
listings_final.head()
```

Recalculating Mortgage

```python
listings_final['mortgage_payout_after_listing'] = listings_final.apply(lambda x: Mortgage(interest=x.mortgage_rate, amount=x.mortgage_amount_after_listing, months=x.mortgage_duration).total_payout(), axis=1)
listings_final['mortgage_payout_after_listing'] = listings_final['mortgage_payout_after_listing'].astype(int)
listings_final.head()
```

Recalculating Interest

```python
listings_final['interest_paid_after_listing'] = listings_final['mortgage_payout_after_listing'] - listings_final['mortgage_amount_after_listing']
listings_final.head()
```

Savings that user could have saved

```python
listings_final['savings_total'] = listings_final['interest_paid'] - listings_final['interest_paid_after_listing']
listings_final.head()
```

```python
listings_final['savings_each_month'] = listings_final['savings_total'] / listings_final['mortgage_duration']
listings_final.head()
```

Creating Investment directory

```python
investment_directory = {}
investor_balance_amount = investors_df.to_dict('records')
id = 1
minimum_investment = investors_df.investment.min
maximum_investment = investors_df.investment.max
minimum_listing_amount = listings_final.listing_amount.min
maximum_percentage = 0.4
for index,row in listings_final.iterrows():
    actual_listing_amount = row['listing_amount'] 
    listing_amount = row['listing_amount'] 
    percentage_remaining = 1
    while percentage_remaining > 0 and listing_amount > 100:
        investor_id = randint(0, investors_df.shape[0] - 1)
        investor_amount = investor_balance_amount[investor_id]['investment']
        percentage_to_buy = randint(5,40) / 100
        if(percentage_remaining - percentage_to_buy > 0):
            percentage_remaining = percentage_remaining - percentage_to_buy
        else:
            percentage_to_buy = percentage_remaining
            percentage_remaining = 0
        purchase_amount = percentage_to_buy * listing_amount
        if(investor_amount >  purchase_amount):
            investor_balance_amount[investor_id] = {'investment' : investor_amount -  purchase_amount}
            actual_listing_amount = listing_amount
            listing_amount = listing_amount - purchase_amount
            investment_directory[id] = row['index'], investor_id, percentage_to_buy, percentage_remaining,  purchase_amount, actual_listing_amount, listing_amount
            id = id + 1
    
    
```

```python
investment_directory_df = pd.DataFrame.from_dict(investment_directory, orient='index', columns=['listing_id','investor_id', 'percentage_to_buy', 'percentage_remaining', 'purchase_amount', 'Last_listing_amount', 'current_listing_amount'])
investment_directory_df.head()
```
